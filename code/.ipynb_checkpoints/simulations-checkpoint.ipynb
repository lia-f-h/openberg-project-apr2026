```python
# Model configurations:
dict_in = {'config':{'seed:wind_form_drag_coef':1.5, 'seed:water_form_drag_coef':1.5,}}

# Input data
dict_in['input']=['cmems_mod_arc_phy_anfc_6km_detided_PT1H-i',{'era5':{'ref':'../input/era5_2023_09_10-18.nc'}}]#,'cmems_mod_arc_phy_my_topaz4_P1D-m','cmems_mod_glo_phy_anfc_0.083deg_PT1H-m','cmems_mod_glo_phy_my_0.083deg_P1D-m']

# Initial information
n=1
randspace=np.random.rand(n)
lengths, widths, sails, drafts = 81, 72, 2.9914802407484546, 21.199466273020537 #lengths, widths = randspace*(87-75)+75, randspace**(80-65)+65
lat, lon = 80.5794, 37.81915 
dict_in['seed']= {'lon': lon, 'lat': lat,
            'length': lengths, 'width': widths, 
            'sail': sails, 'draft': drafts, 
            'number': n, 
           'time': datetime(2023,9,10,12,36)}
dict_in['more']={'duration':0.5,'logfile':50}

o1,o1_ds = simu_late(dict_in) #o1 can be used to create plots. o1_dataset gives back an xr.dataset
```
